

> In the following peice  of code all the hashtags and mentions are extracted and fed in to a new column.



```python
def find_hashtags(tweet):
    #This function extracts hashtags from the tweets.
    return re.findall('(#[A-Za-z]+[A-Za-z0-9-_]+)', tweet) 

def find_mentions(tweet):
    #This function extracts hashtags from the tweets.
    return  re.findall('(@[A-Za-z]+[A-Za-z0-9-_]+)', tweet)
    
df1['hashtags']  = df1.tweets.apply(find_hashtags)
df1['mentions']  = df1.tweets.apply(find_mentions)
#= df1.tweets.apply(find_hashtags)
df1

for i in df_list :
  i['hashtags']  = i.tweets.apply(find_hashtags)
  i['mentions']  = i.tweets.apply(find_mentions)

df2
```
